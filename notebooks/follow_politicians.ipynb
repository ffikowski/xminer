```python
import requests, time

bearer = "USER_CONTEXT_BEARER_WITH_follows.write"
source_user_id = "YOUR_ACCOUNT_USER_ID"
usernames = ["GregorGysi", "ArminLaschet", "meister_schafft"]

# 1) usernames -> IDs (batch up to 100)
r = requests.get(
    "https://api.x.com/2/users/by",
    params={"usernames": ",".join(usernames), "user.fields": "id,username"},
    headers={"Authorization": f"Bearer {bearer}"}
)
r.raise_for_status()
ids = [u["id"] for u in r.json().get("data", [])]

# 2) follow each id
for uid in ids:
    resp = requests.post(
        f"https://api.x.com/2/users/{source_user_id}/following",
        json={"target_user_id": uid},
        headers={"Authorization": f"Bearer {bearer}", "Content-Type": "application/json"}
    )
    print(uid, resp.status_code, resp.text)
    time.sleep(1)
```
